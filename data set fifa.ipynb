```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset FIFA
data = pd.read_csv("fifa20_data.csv", low_memory=False)

# Membersihkan kolom Value
data['Value'] = data['Value'].replace({'€':'', 'M':'e6', 'K':'e3'}, regex=True).map(eval)

# Membersihkan kolom Wage
data['Wage'] = data['Wage'].replace({'€':'', 'M':'e6', 'K':'e3'}, regex=True).map(eval)

# Kolom yang ingin dianalisis
kolom_filter = ['Age', 'Value', 'Wage']

# Loop plotting
for col in kolom_filter:
    print(f"Column: {col}")
    print(f"Skew: {round(data[col].skew(), 2)}")
    
    plt.figure(figsize=(15,4))
    
    # Histogram
    plt.subplot(1,2,1)
    data[col].hist(grid=False)
    plt.ylabel('Count')
    plt.title(f'Histogram of {col}')
    
    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    
    plt.show()

# Menampilkan jumlah posisi pemain
print("\nJumlah Position:")
print(data['Position'].value_counts())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# membaca dataset
data = pd.read_csv("fifa20_data.csv", low_memory=False)

# membersihkan nama kolom
data.columns = data.columns.str.strip()

# setup style biar bersih
sns.set_style("whitegrid")

# membuat figure
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# =========================
# 1. AGE
# =========================
sns.histplot(
    data=data,
    x='Age',
    bins=15,
    ax=axes[0],
    color='blue'
)

axes[0].set_title('Distribusi Umur Pemain')

# =========================
# 2. WAGE
# =========================
if 'Wage' in data.columns:

    # ambil top 15 wage tertinggi
    wage_data = data[['Name', 'Wage']].dropna().head(15)

    sns.barplot(
        data=wage_data,
        x='Wage',
        y='Name',
        ax=axes[1],
        color='green'
    )

    axes[1].set_title('Top Wage Player')

# =========================
# 3. CLUB
# =========================
if 'Club' in data.columns:

    top_club = data['Club'].value_counts().head(10)

    sns.barplot(
        x=top_club.values,
        y=top_club.index,
        ax=axes[2],
        color='red'
    )

    axes[2].set_title('Top 10 Club')

# rapikan layout
plt.tight_layout()

# tampilkan
plt.show()
```
